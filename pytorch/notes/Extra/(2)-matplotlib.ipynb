# How to use Matplotlib

Matplotlib is a powerful Python library for creating static, animated, and interactive visualizations. It is widely used for data visualization tasks in fields such as data science, machine learning, and scientific computing. Matplotlib provides a wide range of functionalities to create various types of plots, including line plots, scatter plots, bar plots, histograms, pie charts, and more.

1. __Importing Matplotlib__: The first step is to import Matplotlib into your Python script or Jupyter Notebook.

```python
import matplotlib.pyplot as plt
```

2. __Basic Line Plot__: Let's start by creating a simple line plot

```python
# Sample Data
x = [1,2,3,4,5]
y = [2,4,6,8,10]

# Plotting the data
plt.plot(x, y)

# Adding plot labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple Line Plot')

# Display the plot
plt.savefig('images/plot1.png')
plt.show()
```

3. __Customizing Plots__: Matplotlib allows you to customize various aspects of your plots such as colors, line styles, markers, gridlines, etc.

```python
plt.plot(x, y, color='red', linestyle='--', marker='o', markersize=8, label='Line 1')
plt.legend() # Show Legend
plt.grid(True) # show grid lines
plt.savefig('images/plot2.png')
```

4. __Multiple Plots__: You can create multiple plots in the same figure using subplot

```python
plt.subplot(2,1,1) # (Rows, columns, plot_number)
plt.plot(x,y)

plt.subplot(2,1,2)
plt.plot(x, [i**2 for i in y], 'r--') # Squaring the y values
plt.savefig('images/plot3.png')
```

5. __Other Plot Types__: Matplotlib supports various plt types like scatter plots, bar plots, histograms, pie charts, etc.

```python
# Scatter Plot
plt.scatter(x, y)
plt.savefig('images/plot4-1.png')

plt.show()

# Bar Plot
plt.bar(x, y)
plt.savefig('images/plot4-2.png')
plt.show()

# Histogram
plt.hist(y, bins=5)
plt.savefig('images/plot4-3.png')
plt.show()


# Pie Chart
labels = ['A','B','C','D']
sizes= [15, 30, 45, 10]
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.savefig('images/plot4-4.png')
plt.show()
```
