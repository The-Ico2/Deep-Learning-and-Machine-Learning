- The mathematical formula for the softmax (Physical Notes)

```python
import numpy as np, torch, torch.nn as nn, matplotlib.pyplot as plt

# manualy in numpy

# the list of numbers
z = [1, 2, 3]

# compute the softmax result
num = np.exp(z)
den = np.sum(np.exp(z))
sigma = num / den

print(sigma)
print(np.sum(sigma))
```

```python
# repeat with some random integers
z = np.random.randint(-5, high=15, size=25)
print(z)

# compute the softmax result
num = np.exp(z)
den = np.sum(num)
sigma = num / den

# Compare
plt.plot(z, sigma, 'ko')
plt.xlabel('Original Numer (z)')
plt.ylabel('Softmaxified $\sigma$')
# plt.yscale('log')
plt.title('$\sum\sigma$ = %g' %np.sum(sigma))
plt.show()
```
