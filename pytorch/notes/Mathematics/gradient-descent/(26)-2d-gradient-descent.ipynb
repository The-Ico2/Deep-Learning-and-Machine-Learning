### Quick note about terminology
- Derivative
    - One dimension
- Partial Derivative
    - Ignoring one dimension and only focusing on the other
    - Gradient
        - The Collection of all the partial derivatives of a fucntion

### Gradient Descent in 2D

```python
# the "peaks" function
def peaks(x,y):
    # espand to a 2D mesh
    x,y = np.meshgrid(x,y)

    z = 3*(1-x)**2 * np.exp(-(x**2) - (y+1)**2) - 10*(x/5 - x**3 - y**5) * np.exp(-x**2-y**2) - 1/3*np.exp(-(x+1)**2 - y**2)
```

```python
# create the linspace
x = np.linspace(-3, 3, 201)
y = np.linspace(-3, 3, 201)

Z = peaks(x,y)

# test plot
plt.imshow(Z,extent=[x[0],x[-1],y[0],y[-1]],vmin=-5,vmax=5,origin='lower')
plt.show()
```
