```python
import math
import torch as t
import matplotlib.pyplot as plt
def ReLU(inp):
    result = max(0,inp)
    print('ReLU:', result)
    return result
def ParaReLU(inp):
    result = max(0.01*inp, inp)
    print('Leaky ReLU:', result)
    return result
def Sigmoid(inp):
    result = 1/(1 + math.exp(-inp))
    print('Sigmoid:', result)
    return result;
def tanh(inp):
    result = (2/(1+math.exp(-2*inp)))-1
    print('tanh:', result)
    return result;

i = t.randn(100)
RLU = t.zeros(i.__len__())
PRLU = t.zeros(i.__len__())
Sig = t.zeros(i.__len__())
Tan = t.zeros(i.__len__())
print(i.__len__())
exp1 = 0
exp2 = 0
exp3 = 0
exp4 = 0
while exp1<i.__len__():
    res = ReLU(i[exp1])
    RLU[exp1] = res
    exp1+=1
while exp2<i.__len__():
    res = ParaReLU(i[exp2])
    PRLU[exp2] = res
    exp2+=1
while exp3<i.__len__():
    res = Sigmoid(i[exp3])
    Sig[exp3] = res
    exp3+=1
while exp4<i.__len__():
    res = tanh(i[exp4])
    Tan[exp4] = res
    exp4+=1

plt.plot(RLU, 'r') # Squaring the y values
plt.plot(PRLU, 'p') # Squaring the y values
plt.plot(Sig, 'b') # Squaring the y values
plt.plot(Tan, 'g') # Squaring the y values


```
